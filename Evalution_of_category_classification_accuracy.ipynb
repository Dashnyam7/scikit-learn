# Evalution of category classification accuracy

## Categorical classification accuracy evaluation index

When evaluating the accuracy of a classification model created for data classification prediction, the indexes handled differ between binary classification and multi-class classification.

There are many indices for evaluating model accuracy, such as precision, recall , F value , and accuracy rate . A table called a mixing matrix is ​​often used to find the four indices mentioned above

image.png

Fit rate: $\frac{tp}{tp+fp}$
Recurrence rate: $\frac{tp}{tp+fn}$
F value: $\frac{2\times fit rate\times reappearance rate}{fit rate + reappearance rate}$
Correct solution rate: $\frac{tp+tn}{tp+fp+fn+tn}$

image.png

## Dataset preperation

Therefore, we extract the data from rows 0 to 100 of the Iris dataset and split it into training and testing datasets.Setosatarget=0Versicolortarget=1

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
iris = load_iris()
data, target = iris.data[:100, :], iris.target[:100]
data_train, data_test, target_train, target_test = train_test_split(data, target, random_state=123, test_size=0.2)
```

## Learning

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(data_train, target_train)
target_pred = tree.predict(data_test)
```

## Evalution of classification accuracy

 let's output an index for evaluating the classification accuracy of the decision tree

```python
from sklearn.metrics import classification_report
print(classification_report(target_test, target_pred))
```

## What is cross validation

 technique called cross-validation is often used to evaluate model accuracy.

Cross-validation is a method of repeatedly dividing a dataset into different training and test datasets to build and evaluate a model

The problem of overfitting can be eliminated by constructing and evaluating the model by changing the split range of the training dataset and the test dataset

Cross-validation is also commonly used to tune model hyperparameters

## Cross validation

k-fold cross-validation

In k-fold cross-validation, the dataset is split into arbitrary numbers, one of which is used as the test dataset and the rest as the training dataset

K-fold cross-validation repeatedly builds and evaluates the model while shifting the division range.

image.png

A k-fold cross-validation with a constant proportion of the test dataset in the dataset is specifically called a stratified k-fold cross-validation . You can easily perform stratified k-fold cross-validation using functions from
scikit-klearn's model_selectionmodule .cross_val_score

```python
from sklearn.model_selection import cross_val_score
```

cv=10specifying to split into 10 pieces of the dataset

scoringby setting parameters, you can specify which indicators to output

```python
cross_val_score(tree, data, target, cv=10, scoring='accuracy')
```
